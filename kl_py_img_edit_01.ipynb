```python
from PIL import Image
from PIL.ExifTags import TAGS

im1 = Image.open(r"c:\Users\kecskemetil\Downloads\_301c1488-cca5-48ef-a1d1-c1fea50a6a0b.jfif");
im2 = Image.open(r"c:\Users\kecskemetil\Pictures\kl_mie.jpg");


def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width , im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0,im1.height -61))
    return dst

get_concat_h(im1, im2).save(r'c:\Users\kecskemetil\Pictures\kl_mie_01.jpg')
im3 = get_concat_h(im1, im2)
im3.show();
```
