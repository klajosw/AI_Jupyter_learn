```python
## Solplavet JSON adatok feldolgozása !! havi összesített !!
import json
import datetime
import pandas as pd
from datetime import datetime

most = datetime.now()

file = 'kl1.json'  ## havi összesített
with open(file) as train_file:
    dict_train = json.load(train_file)

train = pd.DataFrame.from_dict(dict_train, orient='index')
train.reset_index(level=0, inplace=True)

print(train)              ## teljes json
print(train[0][1])        ## csak az aktuális adat sor
t0 = train[0][1]
me = t0['unit']           ## mértékegység
idosor = t0['x']          ## idősor
adatok = t0['data']       ## adatsorok
adatok_ac = adatok['ac']  ## egyen áram érték számsorok
adatok_dc = adatok['dc']  ## váltó áram érték számsorok
print(me)                 ## mértékegység
print(idosor)
print(adatok_ac)          ## egyen áram érték számsorok
print(adatok_dc)          ## váltó áram érték számsorok

print('------1------')

tartomany = range(12)

for i in tartomany:
    print(i, idosor[i ], 'Honap', adatok_dc[i],' MWh |', adatok_ac[i],' MWh |')

display(
    train[0][1]['unit'],           ## direkt mértékegység
    train[0][1]['x']   ,           ## direkt idosor
    train[0][1]['data']['ac'],     ## direkt ac
    train[0][1]['data']['dc'])  ## direkt dc

## összefüzés xls kiíráshoz
df_xls = pd.DataFrame({'IDO': train[0][1]['x'], 
                      'AC' : train[0][1]['data']['ac'],
                      'DC' : train[0][1]['data']['dc']
                      })
## pd.concat([df, series], axis = 1))

## érvényességidő kinyerése
szam = int(train[0][3])
timestamp = szam / 1000
date_time = datetime.fromtimestamp(timestamp)
#datum = date_time.strftime("%Y_%m_%d_%H:%M:%S")
datum = date_time.strftime("%Y_%m_%d")

display(df_xls)

## egy excelbe mentés formázás nélkül
# df_xls.to_excel(r"c:\Users\kecskemetil\Documents\sj\napelem\solplanet_napok\KL_json_evi_ossz.xlsx", sheet_name='Munka1',index=False)

writer = pd.ExcelWriter(r'c:\Users\kecskemetil\Documents\sj\napelem\solplanet_napok\KL_json_havi_ossz_'+datum+'.xlsx')
df_xls.to_excel(writer, sheet_name='Havi_ossz', index=False)

workbook  = writer.book
worksheet = writer.sheets['Havi_ossz']
(max_row, max_col) = df_xls.shape

indexek=[]
indexek.extend(df_xls.columns)

# Tábla formázás
column_settings = [{'header': column} for column in indexek]


worksheet.add_table(0, 0, max_row, max_col - 1, {'columns': column_settings}) ## index nélkül
worksheet.set_column(0, max_col - 1, 15) ## oszlop szélessség 15-re állítás   ## index nélkül

# Az excel állomány mentése
writer.close()
```
